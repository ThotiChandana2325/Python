```python
import numpy as np
#series functions
#map() map column of one series to other series
x=pd.Series(["one","two","three",np.nan])
print(x)
x.map({"one":"a"})
x.map("the number is {}".format,na_action="ignore")#na_action to return null values as nan

#std()calculates standard deviation of data sets
import statistics as s
x=pd.Series([23,54,23,13,89,299])
n=x.to_frame()#series to frame
print(n)
print("standard deviation is:",x.std())
print()
y=pd.DataFrame({
    "Name":["Chandu","Srav","Subbu"],
    "Dsp":[37,38,34],
    "Wt":[39,37,38],
   })
print(y[y.Wt==y["Wt"].max()])
print()
print(y.columns)
print(y.index)
print(y.std(ddof=2))#delta degree of freedom
y
```

```python
x=pd.Series([10,20,32,4,2,12])
y=pd.Series([43,32,44,21,42,7])
df=pd.DataFrame(columns=["num1","num2"])
df["num1"]=x
df["num2"]=y
df["total"]=df.sum(axis=1)
bins=[18,29,55,69,80]
name=["D","C","B","A"]
df["grade"]=pd.cut(df["total"],bins,labels=name)
def highlight(s):
    if s.total<26:
        return ['background_color=red']
df.style.apply(highlight,axis=1)       
df
```
