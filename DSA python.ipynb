# Time Complexity

ORDER OF GROWTH APPRAOCH:

constant    : O(1) for operations without loop,functions without loop and recursion
linear      : O(n) i/p incr,in the same way iteration(time) incr 
quadratic   : O(n^2) i/p inc,time is square of time
logarithmic : O(logn) i/p multiplied,time added (when incre in (//)
exponential : O(2^n) i/p added,iteration multiplies (in form of ^n)
AMONG ALL log(n) is VERY EFFICIENT
