```python
class calculator:
    def add(self,x,y):
        return x+y
    def sub(self,x,y):
        return x-y
    def mul(self,x,y):
        return x*y
    def div(self,x,y):
        return x/y
    def calculate(self,x,y,ope):        
            if ope=="+":
                return self.add(x,y)
            elif ope=="-":
                return self.sub(x,y)
            elif ope=="*":
                return self.mul(x,y)
            elif ope=="/":
                return self.div(x,y)
            else:
                print("invalid operator")
                
cal=calculator()
x=int(input("enter a number"))
result=x
while True:
    y=int(input("enter another number"))
    operator=input("enter operator to operation:/ {0} to exit")
    result=cal.calculate(result,y,operator)
    print("Answer =",result)

    if operator=="0":        
        break
        
```

Class Methods:

    Definition:
        A class method is a method that is bound to the class and not the instance.
        It takes the class itself as its first parameter, conventionally named cls.

    Declaration:
        Decorated with @classmethod.

    Access to Class Variables:
        Can access and modify class-level variables.

    Access to Instance Variables:
        Can't access or modify instance-level variables directly unless provided with an instance.

    Use Case:
        Often used when the method involves operations on class-level attributes or when you need to perform some action that is related to the class itself.

Example:

python

class MyClass:
    class_variable = 0

    @classmethod
    def class_method(cls, x):
        cls.class_variable += x
        print(f"Class variable: {cls.class_variable}")

# Using class method
MyClass.class_method(5)  # Output: Class variable: 5

Static Methods:

    Definition:
        A static method is a method that is bound to the class and not the instance.
        It doesn't take the instance or class as its first parameter.

    Declaration:
        Decorated with @staticmethod.

    Access to Class Variables:
        Can access class-level variables.

    Access to Instance Variables:
        Cannot access or modify instance-level variables directly.

    Use Case:
        Used when a method doesn't need access to instance-specific or class-specific variables. It's essentially a utility function that is logically related to the class.

Example:

python

class MyClass:
    @staticmethod
    def static_method(x):
        print(f"Static method: {x}")

# Using static method
MyClass.static_method(10)  # Output: Static method: 10

In summary, class methods and static methods are both associated with the class rather than instances, but class methods take the class as their first parameter (cls) and are often used for operations on class-level attributes. Static methods, on the other hand, don't take the class or instance as their first parameter and are used for utility functions not directly related to instance or class variables.


Method Overriding in Python:

Method overriding in Python is similar to other object-oriented languages. When a subclass provides a specific implementation for a method that is already defined in its superclass, it's considered method overriding.

Example:

python

class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

# Example usage:
animal = Animal()
animal.sound()  # Output: Animal makes a sound

dog = Dog()
dog.sound()  # Output: Dog barks

In this example, the Dog class overrides the sound method of the Animal class. The overridden method in the subclass provides a specific implementation.
Method Overloading in Python:

Python does not support traditional method overloading with multiple methods having the same name but different parameters as seen in languages like Java. However, Python supports a form of method overloading through default values and variable-length arguments.

Example:

python

class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

# Example usage:
calc = Calculator()
result1 = calc.add(2)
result2 = calc.add(2, 3)
result3 = calc.add(2, 3, 4)

print(result1)  # Output: 2
print(result2)  # Output: 5
print(result3)  # Output: 9

In this example, the add method has default values for b and c, allowing it to be called with different numbers of arguments. While it's not traditional method overloading, it achieves a similar effect by providing flexibility in how the method is called.

In summary, method overriding in Python is similar to other languages, while method overloading is achieved through default values and variable-length arguments, rather than having multiple methods with different signatures.

```python
def wordBreak(s, dictionary):
    n = len(s)
    
    # dp[i] will be True if the first i characters of s can be segmented into words from the dictionary
    dp = [False] * (n + 1)
    dp[0] = True  # Empty string is always breakable
    
    for i in range(1, n + 1):
        for word in dictionary:
            word_len = len(word)
            if i >= word_len and s[i - word_len:i] == word:
                dp[i] = dp[i] or dp[i - word_len]
    
    return 1 if dp[n] else 0

# Example usage:
n1 = 6
s1 = "ilike"
dictionary1 = {"i", "like", "sam", "sung", "samsung", "mobile"}
print(wordBreak(s1, dictionary1))  # Output: 1

n2 = 6
s2 = "ilikesamsung"
dictionary2 = {"i", "like", "sam", "sung", "samsung", "mobile"}
print(wordBreak(s2, dictionary2))  # Output: 1

        
            
```

```python
import os

def search_file_recursive(directory, target_file):
    # Get a list of all files and folders in the current directory
    items = os.listdir(directory)

    for item in items:
        # Construct the full path to the current item
        item_path = os.path.join(directory, item)

        if os.path.isfile(item_path):
            # If the current item is a file, check if it's the target file
            if item == target_file:
                print(f"Found {target_file} at {item_path}")
                return item_path
        elif os.path.isdir(item_path):
            # If the current item is a directory, recursively search within it
            result = search_file_recursive(item_path, target_file)
            if result:
                return result

    # If the target file is not found in the current directory or its subdirectories
    return None

# Example usage
root_directory = input("enter directory path")
target_file = input("enter file name")
result = search_file_recursive(root_directory, target_file)

if result:
    print(f"{target_file} found at: {result}")
else:
    print(f"{target_file} not found in the directory structure.")
```

```python
from cryptography.fernet import Fernet

# we will be encrypting the below string.
message = "hello geeks"

# generate a key for encryption and decryption
# You can use fernet to generate 
# the key or use random key generator
# here I'm using fernet to generate key

key = Fernet.generate_key()

# Instance the Fernet class with the key

fernet = Fernet(key)

# then use the Fernet class instance 
# to encrypt the string string must
# be encoded to byte string before encryption
encMessage = fernet.encrypt(message.encode())

print("original string: ", message)
print("encrypted string: ", encMessage)

# decrypt the encrypted string with the 
# Fernet instance of the key,
# that was used for encrypting the string
# encoded byte string is returned by decrypt method,
# so decode it to string with decode methods
decMessage = fernet.decrypt(encMessage).decode()

print("decrypted string: ", decMessage)
```
